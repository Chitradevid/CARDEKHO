```python
import pandas as pd
import re

# Function to clean and convert 'km' column to numeric
def clean_km(km_value):
    # Convert to string if it's not already
    if not isinstance(km_value, str):
        km_value = str(km_value)
    # Remove non-numeric characters (e.g., "km", "driven")
    cleaned_km = re.sub(r'[^\d]', '', km_value)
    # Convert to numeric
    return pd.to_numeric(cleaned_km, errors='coerce')

# Apply the cleaning function to the 'km' 
df['km'] = df['km'].apply(clean_km)
```

```python
import pandas as pd

# Function to convert price strings to integers
def convert_price_to_int(price_str):
    price_str = price_str.lower().replace(',', '').strip()  # Remove commas and convert to lowercase
    price_str = price_str.replace('₹', '').strip()  # Remove currency symbol

    # Replace 'lakh', 'crore', and 'thousand' with their numeric equivalents
    if 'lakh' in price_str:
        price_num = float(price_str.replace('lakh', '').strip()) * 100000
    elif 'crore' in price_str:
        price_num = float(price_str.replace('crore', '').strip()) * 10000000
    elif 'thousand' in price_str:
        price_num = float(price_str.replace('thousand', '').strip()) * 1000
    else:
        price_num = float(price_str)  # For cases like '58000' without lakh, crore, or thousand
    
    return int(price_num)

# Apply the conversion function to the price column
df['price'] = df['price'].apply(convert_price_to_int)
```

```python
import pandas as pd

# List of columns to process
columns_to_process = ['Gear Box', 'Height', 'Kerb Weight', 'Length', 'Torque', 'Wheel Base', 
                      'Width', 'Mileage',"Turning Radius","Wheel Size","Top Speed","Max Torque",
                     "Rear Tread","Front Tread","Cargo Volumn","Acceleration","Alloy Wheel Size","Max Power",
                     "km","Registration Year","Seats","Engine Displacement"]

for col in columns_to_process:
    # Check if the column is of type object (string)
    if df[col].dtype == 'object':
        # Remove non-numeric characters
        df[col] = df[col].str.extract('(\d+)')  # Extract numeric part
        
        # Convert the column to float to calculate the mean
        df[col] = df[col].astype(float)
        
        #Calculate the mean of the column (ignoring NaN)
        mean_value = df[col].mean()
        
        #Fill NaN values with the mean
        df[col] = df[col].fillna(mean_value)
        
        #Convert the column to integers
        df[col] = df[col].astype(int)

df
```

```python
# Fill missing values for object columns with mode
object_cols = df.select_dtypes(include=['object']).columns
for col in object_cols:
    mode_value = df[col].mode()[0]  # Get the mode of the column
    df[col].fillna(mode_value, inplace=True)

# Fill missing values for int and float columns with mean
numeric_cols = df.select_dtypes(include=['int64', 'float64',"int32"]).columns
for col in numeric_cols:
    mean_value = df[col].mean()  # Get the mean of the column
    df[col].fillna(mean_value, inplace=True)
```
